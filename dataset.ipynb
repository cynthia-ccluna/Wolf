測試用短片

```python
import yt_dlp

# 這是測試用的短影片連結（可以替換為你想要的影片）
url = 'https://www.youtube.com/watch?v=dQw4w9WgXcQ'

# 設定只下載影片的前1分鐘音訊
ydl_opts = {
    'format': 'bestaudio/best',
    'download_sections': ['*00:00-01:00'],  # 只下載前1分鐘
    'outtmpl': '/content/audio/%(title)s.%(ext)s',  # 儲存位置與檔案名稱
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
```

```python
import whisper

# 載入Whisper模型
model = whisper.load_model("tiny")

# 設定檔案名稱（假設剛下載的文件名稱是 "Rick Astley - Never Gonna Give You Up (Official Music Video).mp3"）
audio_file = '/content/audio/Rick Astley - Never Gonna Give You Up (Official Music Video).mp3'

# 進行語音辨識
result = model.transcribe(audio_file)

# 顯示結果
print(result['text'])
```

```python
import whisper
import pandas as pd

model = whisper.load_model("tiny")  # 可改成 "tiny" 做快速測試
result = model.transcribe("clipped_audio.mp3")

# 整理成表格格式
rows = []
for seg in result["segments"]:
    rows.append({
        "start_time": pd.to_datetime(seg["start"], unit="s").strftime("%H:%M:%S"),
        "end_time": pd.to_datetime(seg["end"], unit="s").strftime("%H:%M:%S"),
        "text": seg["text"].strip()
    })

df = pd.DataFrame(rows)

# 顯示表格
df.head()
```

```python
# 安裝必要套件
!pip install -U openai-whisper pandas

# 下載剛剛的 mp3（已經下載好的話直接指定路徑）
AUDIO_FILE = "sample.mp3"

# 讀取 Whisper
import whisper
import pandas as pd

model = whisper.load_model("small")  # small比較快，medium更準

# 轉錄
result = model.transcribe(AUDIO_FILE, word_timestamps=False)

# 整理成 DataFrame
data = []
for segment in result['segments']:
    start = segment['start']
    end = segment['end']
    text = segment['text']
    data.append({
        'start_time': pd.to_datetime(start, unit='s').strftime('%H:%M:%S'),
        'end_time': pd.to_datetime(end, unit='s').strftime('%H:%M:%S'),
        'text': text.strip()
    })

df = pd.DataFrame(data)

# 顯示表格
df.head()

# 輸出成 CSV（可以直接下載開Excel）
df.to_csv("whisper_output.csv", index=False)
```
